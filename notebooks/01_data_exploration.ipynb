```python
import sys
import os

# Add src to path so imports work
sys.path.append(os.path.abspath('..'))

from src.data.loader import NuScenesLoader
from src.data.visuals import create_surround_montage
import matplotlib.pyplot as plt

# Initialize Loader
loader = NuScenesLoader()
```

```python
# Get a random sample
samples = loader.get_all_samples()
sample_token = samples[15] # Arbitrary index

# Get Metadata
desc = loader.get_scene_description(sample_token)
print(f"Scene Description: {desc}")
print(f"Sample Token: {sample_token}")

# Get Paths
paths = loader.get_camera_paths(sample_token)
for cam, path in paths.items():
    print(f"{cam}: {path}")
```

```python
# Generate Montage
montage = create_surround_montage(paths, resize_factor=0.5)

# Display
plt.figure(figsize=(18, 10))
plt.imshow(montage)
plt.axis('off')
plt.title(f"Surround View | {desc}")
plt.show()

# Verify Resolution
print(f"Montage Resolution: {montage.size}")
```
